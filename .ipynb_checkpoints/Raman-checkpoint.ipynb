```python
import numpy as np
import matplotlib.pyplot as plt
in_file = open("041119/aba1_1.txt")
x = np.loadtxt("041119/aba1_1.txt",unpack = True)
print(x)
```

```python
plt.plot(x[0],x[2])
```

```python
s = np.loadtxt("041119/silicon_1.txt",unpack = True)
plt.plot(s[0],s[2])
```

```python
plt.plot(x[0],x[2],label="aba",color="green")
plt.plot(s[0],s[2],'r',label="silicon")
plt.legend()
plt.suptitle("Intensity vs wavelength", fontsize = 14)
plt.ylabel("Intensity")
plt.xlabel("Wavelength (nm)")
fig = plt.gcf()
fig.set_size_inches(9,6)
plt.savefig("silicon_vs_aba.png", dpi=300)
```

```python
q = np.loadtxt("041119/abc_maybe_1_1.txt", unpack = True)
plt.plot(x[0],x[2],label="aba",color="green")
plt.plot(q[0],q[2],'r',label="abc")
plt.legend()
plt.suptitle("Intensity vs wavelength", fontsize = 14)
plt.ylabel("Intensity")
plt.xlabel("Wavelength (nm)")
fig = plt.gcf()
fig.set_size_inches(9,6)
plt.savefig("aba_vs_abcmaybe.png", dpi=300)
```

```python
plt.plot(x[0][740:760],x[2][740:760],label="aba",color="green")
plt.plot(q[0][740:760],q[2][740:760],'r',label="abc")
plt.legend()
plt.suptitle("Intensity vs wavelength", fontsize = 14)
plt.ylabel("Intensity")
plt.xlabel("Wavelength (nm)")
fig = plt.gcf()
fig.set_size_inches(9,6)
plt.savefig("aba_vs_abcmaybe_zoomed.png", dpi=300)
```

```python
## Delta^2 DRV
%pwd
import numpy as np
import matplotlib.pyplot as plt
#in_file = open("041119/aba1_1.txt")
#print(in_file.read())
open("1.txt","r")
x1 = np.loadtxt("041119/aba1_1.txt")
x2 = np.loadtxt("041119/silicon_1.txt")
x5 = np.loadtxt("041119/FINAL_aba_1.txt")
x7 = np.loadtxt("041119/abc_maybe_1_1.txt")

print(len(x1))
x3,x4,x6,x8 = [],[],[],[]
for i in range(740,761):
    x3.append((x1[i][2]-x2[i][2])**2)
    x4.append(x1[i][0])
    x6.append((x1[i][2]-x5[i][2])**2)
    x8.append((x1[i][2]-x7[i][2])**2)
plt.figure()
plt.plot(x4,x3,"r--")
plt.figure()
plt.plot(x4,x6,"r--")
plt.figure()
plt.plot(x4,x8,"r--")
plt.show()
print(x8)
```
