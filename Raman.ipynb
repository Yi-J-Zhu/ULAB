```python
import numpy as np
import matplotlib.pyplot as plt

folder = "042519/"
suffix = ".txt"

filenames = ["aba_1_1",  "aba_2_1",  "aba_3_1",  "abc_1_1",  "abc_2_1",  "abc_3_1"]
files = [np.loadtxt(folder + file +suffix, unpack = True) for file in filenames]
```

```python
avg_aba = (np.array(files[0]) + np.array(files[1]) + np.array(files[2]) ) / 3.0
print (np.shape(avg_aba))

fig = plt.figure()
plt.plot(avg_aba[0], avg_aba[2])

#plt.plot(files[0][0], files[0][2], 'r')
#plt.plot(files[1][0], files[1][2], 'g')
#plt.plot(files[2][0], files[2][2], 'm')

fig.suptitle('Average ABA', fontsize=20)

fig.savefig('avg_aba.jpg')
```

```python
avg_abc = (np.array(files[3]) + np.array(files[4]) + np.array(files[5]) ) / 3.0
print (np.shape(avg_aba))

fig = plt.figure()
plt.plot(avg_abc[0], avg_abc[2])

#plt.plot(files[3][0], files[3][2], 'r')
#plt.plot(files[4][0], files[4][2], 'g')
#plt.plot(files[5][0], files[5][2], 'm')

fig.suptitle('Average ABC', fontsize=20)

fig.savefig('avg_abc.jpg')
```

```python
diff = avg_aba-avg_abc

fig = plt.figure()
plt.plot(diff[0], diff[2])
fig.suptitle('Difference', fontsize=20)

fig.savefig('aba_abc_diff.jpg')
```
