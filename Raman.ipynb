```python
import numpy as np
import matplotlib.pyplot as plt

minIndex = 585
maxIndex = 595

folder = "042519/"
suffix = ".txt"

filenames = ["aba_1_1",  "aba_2_1",  "aba_3_1",  "abc_1_1",  "abc_2_1",  "abc_3_1"]
files = [np.loadtxt(folder + file +suffix, unpack = True) for file in filenames]
```

```python
avg_aba = (np.array(files[0]) + np.array(files[1]) + np.array(files[2]) ) / 3.0
print (np.shape(avg_aba))

fig = plt.figure()
plt.plot(avg_aba[0][minIndex:maxIndex], avg_aba[2][minIndex:maxIndex], label='average')

plt.plot(files[0][0][minIndex:maxIndex], files[0][2][minIndex:maxIndex], 'r', label='1_1')
plt.plot(files[1][0][minIndex:maxIndex], files[1][2][minIndex:maxIndex], 'g', label='2_1')
plt.plot(files[2][0][minIndex:maxIndex], files[2][2][minIndex:maxIndex], 'm', label='3_1')

fig.suptitle('Average ABA', fontsize=20)
plt.xlabel('Wavelength', fontsize=18)
plt.ylabel('Intensity', fontsize=16)
plt.legend()

fig.savefig('zoomed_avg_aba.jpg')
```

```python
avg_abc = (np.array(files[3]) + np.array(files[4]) + np.array(files[5]) ) / 3.0
print (np.shape(avg_aba))

fig = plt.figure()
plt.plot(avg_abc[0][minIndex:maxIndex], avg_abc[2][minIndex:maxIndex], label='average')

plt.plot(files[3][0][minIndex:maxIndex], files[3][2][minIndex:maxIndex], 'r', label='1_1')
plt.plot(files[4][0][minIndex:maxIndex], files[4][2][minIndex:maxIndex], 'g', label='2_1')
plt.plot(files[5][0][minIndex:maxIndex], files[5][2][minIndex:maxIndex], 'm', label='3_1')

fig.suptitle('Average ABC', fontsize=20)
plt.xlabel('Wavelength', fontsize=18)
plt.ylabel('Intensity', fontsize=16)
plt.legend()

fig.savefig('zoomed_avg_abc.jpg')
```

```python
fig = plt.figure()
plt.plot(avg_abc[0][minIndex:maxIndex], avg_abc[2][minIndex:maxIndex], label='abc')
plt.plot(avg_aba[0][minIndex:maxIndex], avg_aba[2][minIndex:maxIndex], label='aba')

fig.suptitle('ABA V ABC', fontsize=20)
plt.xlabel('Wavelength', fontsize=18)
plt.ylabel('Intensity', fontsize=16)
plt.legend()

fig.savefig('aba_abc.jpg')
```

```python
diff = avg_aba-avg_abc

fig = plt.figure()
plt.plot(diff[0][minIndex:maxIndex], diff[2][minIndex:maxIndex])
fig.suptitle('Difference', fontsize=20)
plt.xlabel('Wavelength', fontsize=18)
plt.ylabel('Intensity', fontsize=16)
fig.savefig('aba_abc_diff.jpg')
```
