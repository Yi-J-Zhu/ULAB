```python
import numpy as np
import matplotlib.pyplot as plt

minIndex = 812
maxIndex = 837

folder = "042519/"
suffix = ".txt"

filenames = ["aba_1_1",  "aba_2_1",  "aba_3_1",  "abc_1_1",  "abc_2_1",  "abc_3_1"]
files = [np.loadtxt(folder + file +suffix, unpack = True) for file in filenames]
```

```python
avg_aba = (np.array(files[0]) + np.array(files[1]) + np.array(files[2]) ) / 3.0
print (np.shape(avg_aba))

fig = plt.figure()
plt.plot(avg_aba[0][minIndex:maxIndex], avg_aba[2][minIndex:maxIndex], label='average')

plt.plot(files[0][0][minIndex:maxIndex], files[0][2][minIndex:maxIndex], 'r', label='sample 1')
plt.plot(files[1][0][minIndex:maxIndex], files[1][2][minIndex:maxIndex], 'g', label='sample 2')
plt.plot(files[2][0][minIndex:maxIndex], files[2][2][minIndex:maxIndex], 'm', label='sample 3')

#plt.axvline(1520, color='b', linestyle='--')
#plt.axvline(1540, color='b', linestyle='--')

fig.suptitle('Average ABA*', fontsize=20)
plt.xlabel('Raman Shift ($cm^{-1}$)', fontsize=18)
plt.ylabel('Intensity (a.u.)', fontsize=16)
plt.legend()

fig.set_size_inches(9,6)
#fig.savefig('zoomed_avg_aba.jpg', dpi=300)
```

```python
resids = [files[index][2] - avg_aba[2] for index in range(3)]

fig = plt.figure()

plt.plot(avg_aba[0][minIndex:maxIndex], resids[0][minIndex:maxIndex], 'r', label='sample 1')
plt.plot(avg_aba[0][minIndex:maxIndex], resids[1][minIndex:maxIndex], 'g', label='sample 2')
plt.plot(avg_aba[0][minIndex:maxIndex], resids[2][minIndex:maxIndex], 'm', label='sample 3')

#plt.axvline(1520, color='b', linestyle='--')
#plt.axvline(1540, color='b', linestyle='--')

fig.suptitle('ABA* Residuals', fontsize=20)
plt.xlabel('Raman Shift ($cm^{-1}$)', fontsize=18)
plt.ylabel('Intensity (a.u.)', fontsize=16)
plt.legend()

fig.set_size_inches(9,6)
#fig.savefig('zoomed_aba_residuals_vert_lines.jpg', dpi = 300)
```

```python
avg_abc = (np.array(files[3]) + np.array(files[4]) + np.array(files[5]) ) / 3.0
print (np.shape(avg_aba))

fig = plt.figure()
plt.plot(avg_abc[0][minIndex:maxIndex], avg_abc[2][minIndex:maxIndex], label='average')

plt.plot(files[3][0][minIndex:maxIndex], files[3][2][minIndex:maxIndex], 'r', label='sample 1')
plt.plot(files[4][0][minIndex:maxIndex], files[4][2][minIndex:maxIndex], 'g', label='sample 2')
plt.plot(files[5][0][minIndex:maxIndex], files[5][2][minIndex:maxIndex], 'm', label='sample 3')

#plt.axvline(1520, color='b', linestyle='--')
#plt.axvline(1540, color='b', linestyle='--')

fig.suptitle('Average ABC*', fontsize=20)
plt.xlabel('Raman Shift ($cm^{-1}$)', fontsize=18)
plt.ylabel('Intensity (a.u.)', fontsize=16)
plt.legend()

fig.set_size_inches(9,6)
#fig.savefig('zoomed_avg_abc.jpg', dpi=300)
```

```python
resids2 = [files[index + 3][2] - avg_abc[2] for index in range(3)]

fig = plt.figure()

plt.plot(avg_aba[0][minIndex:maxIndex], resids2[0][minIndex:maxIndex], 'r', label='sample 1')
plt.plot(avg_aba[0][minIndex:maxIndex], resids2[1][minIndex:maxIndex], 'g', label='sample 2')
plt.plot(avg_aba[0][minIndex:maxIndex], resids2[2][minIndex:maxIndex], 'm', label='sample 3')

plt.axvline(1520, color='b', linestyle='--')
plt.axvline(1540, color='b', linestyle='--')

fig.suptitle('ABC* Residuals', fontsize=20)
plt.xlabel('Raman Shift ($cm^{-1}$)', fontsize=18)
plt.ylabel('Intensity (a.u.)', fontsize=16)
plt.legend()

fig.set_size_inches(9,6)
#fig.savefig('zoomed_abc_residuals_vert_lines.jpg', dpi = 300)
```

```python
fig = plt.figure()
plt.plot(avg_abc[0][minIndex:maxIndex], avg_abc[2][minIndex:maxIndex], label='abc')
plt.plot(avg_aba[0][minIndex:maxIndex], avg_aba[2][minIndex:maxIndex], label='aba')

#plt.axvline(1520, color='b', linestyle='--')
#plt.axvline(1540, color='b', linestyle='--')

fig.suptitle('ABA* V ABC*', fontsize=20)
plt.xlabel('Raman Shift ($cm^{-1}$)', fontsize=18)
plt.ylabel('Intensity (a.u.)', fontsize=16)
plt.legend()

fig.set_size_inches(9,6)
#fig.savefig('zoomed_aba_abc.jpg', dpi=300)
```

```python
diff = avg_abc[2] - avg_aba[2]

fig = plt.figure()
plt.plot(avg_aba[0][minIndex:maxIndex], diff[minIndex:maxIndex], label='abc* - aba*')

plt.axvline(1520, color='b', linestyle='--')
plt.axvline(1540, color='b', linestyle='--')

fig.suptitle('Residual Plot', fontsize=20)
plt.xlabel('Raman Shift ($cm^{-1}$)', fontsize=18)
plt.ylabel('Intensity', fontsize=16)
plt.legend()

fig.set_size_inches(9,6)
#fig.savefig('zoomed_aba_abc_resid_vert_lines.jpg', dpi = 300)
```
